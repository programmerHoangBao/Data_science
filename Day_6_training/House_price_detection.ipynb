# House Price Prediction Project
### 1. Problem Definition: 
- Goal: It is your job to predict the sales price for each house. For each Id in the test set, you must predict the value of the SalePrice variable.
- link: https://www.kaggle.com/competitions/home-data-for-ml-course/overview/description
### 2. Feature selection:
- Choose features to train ML Model
- Need to use 'Feature Engineering" to identify Features needed
### 3. Spliting the datasets
#### 3.1 dataset --> X, y
- 'data': dataset
- 'x': data[features]
- 'y': target variable SalePrice
#### 3.2 X, y --> X_train, y_train, X_valid, y_valid

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./data/train.csv")
data.head()
```

```python
data = pd.read_csv("./data/train.csv", index_col = "Id")
data.head()
```

```python
data.columns
```

## 2. Feature Selections

```python
features = ["LotArea", "YearBuilt", "1stFlrSF", "2ndFlrSF", "FullBath", "BedroomAbvGr", "TotRmsAbvGrd"]
```

## 3. Splitting dataset into X and y

```python
X = data[features]
y = data["SalePrice"]
```

```python
X.head()
```

```python
y.head()
```

## X_train, y_train, X_valid, y_valid

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size = 0.8, test_size = 0.2, random_state = 0)
```

```python
X_train.shape
```

```python
X.shape
```

```python
data.shape
```

```python
X_valid.shape
```

### 4. Trainning Machine Learning model

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state = 1)
```

```python
#Fit training data into model
dt_model.fit(X_train, y_train)
```

```python
y_preds = dt_model.predict(X_valid.head())
```

```python
y_preds
```

```python
pd.DataFrame({'y':y_valid.head(), 'y_preds': y_preds})
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(X_train, y_train)
```

```python
rf_val_preds = rf_model.predict(X_valid)
```
