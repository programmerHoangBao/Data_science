### Project: Data Visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv("./data/california_cities.csv")
cities.head()
```

```python
cities["area_total_km2"].sort_values(ascending=False)
```

```python
#Extract latd = "Vĩ độ" and Longd = "Kinh độ"
lat, lon = cities.latd, cities.longd
population, area = cities.population_total, cities.area_total_km2
```

```python
#plot using pyplot 'API
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(8, 6))
plt.title("California city: Population and Area Distribution")
plt.scatter(lat, lon,
           c = np.log10(population), cmap="viridis",
           s = area, linewidths=0, alpha=0.5)
plt.axis("equal")
plt.xlabel("latitude")
plt.ylabel("longlatitude")
plt.colorbar(label = 'log$_{10}$(population)')
plt.clim(3, 7)

#Create a legend for cities's size
area_range = [50, 100, 300, 500]
for area in area_range:
    plt.scatter([], [], s=area, c = 'k', alpha=0.4,
               label = str(area)+"km$^2$")
plt.legend(scatterpoints=1, labelspacing=1, title="City Area")
```
